```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from pymatgen.analysis.diffraction.core import AbstractDiffractionPatternCalculator
```

```python
with MPRester("OagdrlTwLXSmV3tf") as m:
    
    # Structure for material id
    structure = m.get_structure_by_material_id("mp-1023924")
```

```python
#Pattern de diffraction

rInc = XRDCalculator(wavelength='CuKa')
p = rInc.get_pattern(structure)
print(p)
```

Pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å, on va identifier des pics du diffractogramme associé a leur indice de Laue.
Avec le Diffraction pattern on a pour les thetas suivants des valeurs d'intensité qui sont représentés sur le graphique en dessous ainsi que leur indice hkl : 

- Theta 1 = 5.06721588 | I1 = 1.00000000e+02 | hkl = (0,0,0,1)
- Theta 2 = 10.14437419 | I2 = 1.53625599e+01 | hkl = (0,0,0,2)
- Theta 3 = 15.24159435 | I2 = 2.98372773e+00 | hkl = (0,0,0,3)

On retrouve des indices de Laue à quatre termes ce que j'avoue ne pas réellement bien comprendre, en recherchant un peu j'ai fais l'hypothèse que cela venait du fait que des pics de diffractogramme apparaissaient plusieurs fois pour les même hkl et que le quatrième chiffre servait à les différencier car on voit plus loin dans les valeurs de theta plus élevée que les indices hkl change de (0,0,0)
